```python
from IPython.core.display import Image
from PIL import Image
import torch
import pandas as pd

model = torch.hub.load('ultralytics/yolov5', 'custom', '/content/best.pt')  # custom trained model

# Images
im = '/content/test1.jpg'  # or file, Path, URL, PIL, OpenCV, numpy, list

#Resize
im = Image.open(im)
im = im.resize((640,640))

# Inference
results = model(im)

# Results image
results.print()

# Results text
result = results.pandas().xyxy[0]
img_result = result['name']
img_result = pd.DataFrame(img_result)
img_result.value_counts()

img_val = pd.DataFrame(img_result.value_counts())
img_val = img_val.transpose()
img_result_2 = img_result['name'].str.slice(start=0, stop=2)

if len(img_val.columns) > 1:
    if img_result_2[0] == 'un':
        print('Your plants are unhealthy.')
    elif img_result_2[1] == 'un':
        print('Your plants are unhealthy.')
    else:
        print('Your plants are healthy.')
else:
     if img_result_2[0] == 'un':
        print('Your plants are unhealthy.')
     else:
         print('Your plants are healthy.')
```
